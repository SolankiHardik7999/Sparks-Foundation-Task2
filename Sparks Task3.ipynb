# Task:3 Decision Tree

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.tree import plot_tree
import seaborn as sns
sns.set()
```

# Reading Data

```python
data=pd.read_csv('Iris.csv')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
x1=data.drop('Species',axis=1)
x=x1.drop('Id',axis=1)
y=data['Species']
```

```python
sns.boxplot(data=x, orient="h", palette="Set2")
```

# Training and Testing Data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
x_train.shape, y_train.shape
```

# Prediction

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
dt
```

```python
prediction=dt.predict(x_test)
prediction
```

```python
pre_data=x_test.copy()
pre_data.reset_index(drop=True,inplace=True)
pre_data['Predicted']=prediction
pre_data
```

# Confusion Matrix

```python
print("confusion_matrix :-\n",confusion_matrix(y_test,prediction))
print("\nclassification_report :- \n",classification_report(y_test,prediction))
```

# Visualisation

```python
plt.figure(figsize=(25,15))
tree_image=plot_tree(dt,feature_names=x.values,class_names=y.values,filled=True,rounded=True,label='all',precision=4)
plt.savefig('ShowTree')
```
